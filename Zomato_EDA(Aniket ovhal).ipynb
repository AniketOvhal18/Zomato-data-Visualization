```python
df1=df[['name','online_order','book_table','rate','votes','location','rest_type','dish_liked','cuisines','approx_cost(for two people)']]
```

```python
df1.dropna(inplace=True)
```

```python
df1["book_table"].unique()
```

```python
df1['rate'][0].split('/')[0]
```

```python
def rate_cleaning(val):
    if val == 'NEW' or val == '-':
        return 0.0
    else:
        return float(val.split('/')[0])
```

```python
df1['rate'] = df1['rate'].apply(rate_cleaning)
```
